```python
import pandas as pd 
```

```python
# user input
freq = 15 #in minutes

#Value used to calculate mean values over the provided timeframe
resample_str = str(freq) + 'T'

#Search depth level of data relative to the Brick model
search_depth = 1

forecasting_horizon = 4
```

```python
#Pull data from CSV
inital_df = pd.read_csv('bldg1.csv', header=None)
#inital_df.head()
```

```python
inital_df.shape
```

```python
#Format data so is indexed by time and in column format
pivot_data = inital_df.pivot(index=[0], values= [1], columns=[2])
```

```python
pivot_data.head()
```

```python
#Convert time to date time
time_indexed = pivot_data.set_index(pd.to_datetime(pivot_data.index))
time_indexed.head()
```

```python
#Resample the data with the provided user resample time
time_indexed.resample(resample_str).mean()
time_indexed.head(100)
```

```python
time_indexed.shape
```

```python
#List of all attributes in data set 
list(time_indexed.columns)
```

```python
#pulling out the room air temp data. This will act as the target for the AutoML package. 
time_indexed[1]['bldg1.ZONE.AHU02.RM112.Zone Air Temp'].head()
```

```python
#create lists of column names for each level to pull out to create data sets. The levels correspond to the level of the graph relative 
#to the target value. 

level_1 = ['bldg1.ZONE.AHU02.RM112.Zone Supply Air Flow',
	'bldg1.ZONE.AHU02.RM112.Zone Air Temp Setpoint',
	'bldg1.ZONE.AHU02.RM112.Zone Supply Air Temp',
	'bldg1.ZONE.AHU02.RM112.Zone Percent Air Flow']

level_2 = ['bldg1.AHU.AHU02.Supply Air Pressure',
	'bldg1.AHU.AHU02.Cooling Valve Output',
	'bldg1.AHU.AHU01.Outside Air Temp',
	'bldg1.AHU.AHU02.Supply Air Temp',
	'bldg1.AHU.AHU02.Return Air Temp']
	
level_3 = ['bldg1.CHW.Pump1_Start/Stop',
	'bldg1.CHW.ECONOMIZER',
	'bldg1.CHW.Loop Chilled Water Return Temp',
	'bldg1.CHW.Loop Chilled Water Supply Temp']

```

```python
#Create DF of the attribute data for each of the levels
testdata_level_1 = time_indexed[1][level_1]
testdata_level_2 = time_indexed[1][level_1+level_2]
testdata_level_3 = time_indexed[1][level_1+level_2+level_3]

feature_level = [testdata_level_1, testdata_level_2, testdata_level_3]

testdata_level_3.head()
print(test_level_3.shape)
```

```python
#targets is Pd series. Used 
targets = time_indexed[1]['bldg1.ZONE.AHU02.RM112.Zone Air Temp']

#make features array. If the search_depth is 0, then it will be a uni-varient model
if(search_depth != 0): 
 features = feature_level[search_depth-1]
else:
    features = []
```

```python
#remove the last number equal to forecasting horizon for testing
y_train = [targets[: -forecasting_horizon]]
y_test = [targets[-forecasting_horizon:]]

#if depth is 0, will be a uni-varient model, thus no features
if(search_depth != 0):
    X_train = [features[: -forecasting_horizon]]
else:
    X_train = []

X_test = [features[-forecasting_horizon:]]

known_future_features = list(features.columns)
```

```python
#set start time
start_times = [targets.index[0]]
```

```python
#test cell used to analyze the created arrays to be used in the APi call 
from numpy.core.fromnumeric import shape
print("X")
print(type(X_train))
print(shape(X_train))
#print(X_train)
print(type(X_test))
#print(X_test)
print(shape(X_test))

print("Y")
print(type(y_train))
#print(y_train)
print(shape(y_train))
print(type(y_test))
print(y_test)
print(shape(y_test))


print(known_future_features)
print(forecasting_horizon)
print(resample_str)
print(start_times)
print(X_train)
```

```python
from autoPyTorch.api.time_series_forecasting import TimeSeriesForecastingTask
import os
import tempfile as tmp
import warnings
import copy

os.environ['JOBLIB_TEMP_FOLDER'] = tmp.gettempdir()
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'

warnings.simplefilter(action='ignore', category=UserWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)


api = TimeSeriesForecastingTask()

api.search(
    X_train=X_train,
    y_train=copy.deepcopy(y_train),
    X_test=X_test,
    optimize_metric='mean_MASE_forecasting',
    n_prediction_steps=forecasting_horizon,
    memory_limit=16 * 1024,   # Currently, forecasting models use much more memories
    freq=resample_str,
    start_times=start_times,
    func_eval_time_limit_secs=50,
    total_walltime_limit=500,
    min_num_test_instances=1000,  # proxy validation sets. This only works for the tasks with more than 1000 series
    known_future_features=known_future_features,
)



  
```

```python
from autoPyTorch.datasets.time_series_dataset import TimeSeriesSequence

test_sets = []

# We could construct test sets from scratch
for feature, future_feature, target, start_time in zip(X_train, X_test,y_train, start_times):
    test_sets.append(
        TimeSeriesSequence(X=feature.values,
                           Y=target.values,
                           X_test=future_feature.values,
                           start_time=start_time,
                           is_test_set=True,
                           # additional information required to construct a new time series sequence
                           **api.dataset.sequences_builder_kwargs
                           )
    )
```

```python
pred = api.predict(test_sets)
```
